### ARCHIVO 4: Entreno modelo YOLOv11 versiÃ³n segmentaciÃ³n con de dataset de coches segmentados de YOLO

CÃ³digo que entrena el modelo YOLOv11seg dedicado a la segmentaciÃ³n de objetos con imÃ¡genes de coches. 

CÃ³digo ejecutado en plataforma Google Colabs. Se precisa una carga de las imÃ¡genes en el entorno local Colab para un entreno eficiente

```python
# Llibreries

import shutil
import os
from ultralytics import YOLO
from google.colab import drive
```

```python
# Es munta el Drive
drive.mount('/content/drive', force_remount=True)

# Es creen carpetes locals
os.makedirs('/content/data/images/train', exist_ok=True)
os.makedirs('/content/data/labels/train', exist_ok=True)
os.makedirs('/content/data/images/test', exist_ok=True)
os.makedirs('/content/data/labels/test', exist_ok=True)

# Copiar imatges i etiquetes des de Google Drive
shutil.copytree('/content/drive/MyDrive/Colab Notebooks/CarSeg/train/images',
                '/content/data/images/train', dirs_exist_ok=True)

shutil.copytree('/content/drive/MyDrive/Colab Notebooks/CarSeg/train/labels',
                '/content/data/labels/train', dirs_exist_ok=True)

shutil.copytree('/content/drive/MyDrive/Colab Notebooks/CarSeg/test/images',
                '/content/data/images/test', dirs_exist_ok=True)

shutil.copytree('/content/drive/MyDrive/Colab Notebooks/CarSeg/test/labels',
                '/content/data/labels/test', dirs_exist_ok=True)
```

```python
# Carregar model preentrenat YOLOv11n-seg versiÃ³ especÃ­fica de segmentaciÃ³
model = YOLO("yolo11n-seg.pt")

# Entrament model segons YOLO
model.train(
    data="/content/drive/MyDrive/Colab Notebooks/CarSeg/data.yaml",
    epochs=100,
    imgsz=640,
    device='cuda:0',
    batch=4,
    workers=8,
    cache=False,
    project="/content/drive/MyDrive/my_models"  # Dir guardar model
)

# After training, you can validate the model's performance on the validation set
#results = model.val()

# Or perform prediction on new images or videos
#results = model.predict("path/to/your/image.jpg")
```
